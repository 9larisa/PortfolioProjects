```python
# Import libraries

import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12, 8) # Adjusts the configuration of the plots we will create



# Read in the data

df = pd.read_csv('Downloads//movies.csv')


```

```python
# Let's look at the dataabs

df.head()
```

```python
# Let's see it there is any missing data
df = df.dropna()
for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print(f'{col} - {pct_missing}%')
```

```python
# Data types for our columns

df.dtypes
```

```python
# Change data type of columns

df['budget'] = df['budget'].astype('int64')


df['gross'] = df['gross'].astype('int64')
```

```python
df.head()
```

```python
# Create correct Year column

df['year_correct'] = df['released'].str.extract(pat='([0-9]{4})').astype(int)
df.head()
```

```python
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
pd.set_option('display.max_rows', None)
```

```python
# Drop any duplicats

df.drop_duplicates().head()
```

```python
df.head()
```

```python
# Budget high correlation
# company high correlation
```

```python
# Scatter plot with budget vs ross

plt.scatter(x=df['budget'], y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Gross Earnings')
plt.ylabel('Budget for Film')
plt.show()







```

```python
# Plot budget vs gross using seaborn

sns.regplot(x='budget', y='gross', data=df, scatter_kws={'color':'red'}, line_kws={'color':'blue'})
```

```python
# Let's start looking at correlation
```

```python
df.corr(numeric_only=True, method='pearson') #pearson, kendall, spearman
```

```python
# High correlation between budget and gross
```

```python
correlation_matrix = df.corr(numeric_only=True, method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
# Looks at company

df.head()
```

```python
df_numerized = df

for col_name in df_numerized.columns:
    if(df_numerized[col_name].dtype == 'object'):
        df_numerized[col_name] = df_numerized[col_name].astype('category')
        df_numerized[col_name] = df_numerized[col_name].cat.codes
        
df_numerized.head()
```

```python
correlation_matrix = df_numerized.corr(numeric_only=True, method='pearson')

sns.heatmap(correlation_matrix, annot=True)

plt.title('Correlation Matrix for Numeric Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')

plt.show()
```

```python
df_numerized.corr(numeric_only=True)
```

```python
correlation_mat = df_numerized.corr(numeric_only=True)

corr_pairs = correlation_mat.unstack()

corr_pairs
```

```python
sorted_pairs = corr_pairs.sort_values()
sorted_pairs
```

```python
high_corr = sorted_pairs[(sorted_pairs) > 0.5]
high_corr 
```
